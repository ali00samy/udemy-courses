```python
udemy_course['published_timestamp'].str.split('T')
```

```python
udemy_course['published_date'] = udemy_course['published_timestamp'].str.split('T').str.get(0)
```

```python
udemy_course['published_date'] = pd.to_datetime(udemy_course['published_date'], format= '%Y-%M-%d' )
```

```python
udemy_course['published_date'] = udemy_course['published_date'].str.replace('3 hours', '2017-01-18')
```

```python
udemy_course.head()
```

```python
udemy_course['year'] = udemy_course['published_date'].dt.year
```

```python
udemy_course.groupby('year')['num_subscribers'].sum()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data = udemy_course,x='year', y='num_subscribers', hue = 'subject',ci=None)
plt.show()
```
